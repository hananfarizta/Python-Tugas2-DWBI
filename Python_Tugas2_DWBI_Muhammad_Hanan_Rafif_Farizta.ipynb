## Import Library Pandas

```python
import pandas as pd
```

## Baca data dari csv

```python
df = pd.read_csv("/content/titanic.csv")
```

## Menampilkan data csv

```python
df.head()
```

## Membuat column baru (Jumlah Keluarga) berisi value dari SibSp + Parch

```python
df["Jumlah Keluarga"] = df["SibSp"] + df["Parch"]
```

```python
df.head()
```

## Ubah isi column Embarked dari 'S', 'C', 'Q' menjadi 1, 2, 3

```python
categorical_num = {"Embarked" : {"S":1, "C":2, "Q":3}}
df = df.replace(categorical_num)
```

```python
df.head()
```

## Sorting data di column Pclass

```python
df.sort_values(["Pclass"],axis=0, ascending=True,inplace=True,na_position='first')
```

```python
df.head()
```

## Cari data Age yang NaN

```python
df[df['Age'].isnull()].head()
```

```python
df.info()
```

## Imputasi nilai dari kolom age yang kosong (NaN) berdasarkan Pclass

```python
df.dropna(subset=['Age'], inplace=True)
```

```python
df['Age'] = df['Age'].fillna(df.groupby('Pclass')['Age'].transform('mean'))
```

```python
df.groupby('Pclass')['Age'].mean()
```

```python
df.info()
```
